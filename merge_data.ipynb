```python
import pandas as pd
```

```python
path = "D:/Program/dataset/KDD2018/KDD_dataset/process_data/"
bj_aq = pd.read_csv(path+"bj_17_18_aq_cleaned.csv")
bj_aq.head(10)
```

```python
bj_meo = pd.read_csv(path+"bj_17_18_meo_clean.csv")
bj_meo.head(10)
```

```python
# bj_aq "stationId" --->  "station_id"
bj_aq = bj_aq.rename(columns={'stationId':'station_id'})
bj_aq.head()
```

```python
# remove the suffix of station_id (aq and meo)
bj_aq['station_id'] = bj_aq['station_id'].map(lambda x : str(x)[:-3])
bj_meo['station_id'] = bj_meo['station_id'].map(lambda x : str(x)[:-4])
bj_meo.head()
```

```python
bj_meo['station_id'].value_counts()
```

```python
bj_aq['station_id'].value_counts()
```

```python
meo_set = set(bj_meo['station_id'].value_counts().to_dict().keys())
meo_set
```

```python
aq_set = set(bj_aq['station_id'].value_counts().to_dict().keys())
aq_set
```

```python
print('number os station in meo_data: {}'.format(len(meo_set)))
print('number os station in aq_data: {}'.format(len(aq_set)))
```

```python
print('meo_set & aq_set  number : {}'.format(len(meo_set & aq_set)))
meo_set & aq_set
```

```python
def merge_dataframe(meo_data, aq_data, station):
    sta_meo = meo_data[meo_data['station_id'] == station]
    sta_aq = aq_data[aq_data['station_id'] == station].drop(columns='station_id')
#     meo_name = station + '_meo.csv'
#     aq_name = station + '_aq.csv'
#     sta_meo.to_csv(meo_name, index=False)
#     sta_aq.to_csv(aq_name, index=False)
#     sta_meo = pd.read_csv(meo_name)
#     sta_aq = pd.read_csv(aq_name)
    sta = pd.merge(sta_meo, sta_aq, on='utc_time')
    sta.to_csv(path+'merge/'+station + '.csv', index=False)
    print('len of {} is {}'.format(station, len(sta)))
```

```python
for i in (meo_set & aq_set):
    merge_dataframe(bj_meo,bj_aq,i)
```
