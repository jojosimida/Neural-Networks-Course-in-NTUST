```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
```

```python
path = "D:/Program/dataset/KDD2018/KDD_dataset/"
```

```python
bj_aq = pd.read_csv(path+"beijing_17_18_aq.csv")
print(bj_aq.shape)
bj_aq.head()
```

```python
bj_meo = pd.read_csv(path+"beijing_17_18_meo.csv")
print(bj_meo.shape)
bj_meo.head()
```

```python
bj_aq_cols = ['PM2.5', 'PM10', 'NO2', 'CO', 'O3', 'SO2']

bj_aq[bj_aq_cols].hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
              xlabelsize=8, ylabelsize=8, grid=False)  
plt.tight_layout(rect=(0, 0, 1.2, 1.2))

print(bj_aq[bj_aq_cols].describe())
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr =bj_aq[bj_aq_cols].corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Beijing airquality Correlation Heatmap', fontsize=14)
```

```python
bj_meo_cols = ['temperature','pressure', 'humidity', 'wind_direction', 'wind_speed', 'weather']

bj_meo[bj_meo_cols].hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
              xlabelsize=8, ylabelsize=8, grid=False)  
plt.tight_layout(rect=(0, 0, 1.2, 1.2))

print(bj_meo[bj_meo_cols].describe())
```

```python
bj_meo_c = bj_meo[bj_meo_cols]
bj_meo_c.head()
```

```python
for col in bj_meo_c.columns:
    bj_meo_c.loc[bj_meo_c[col] ==999999.000000, col] = np.nan
```

```python
bj_meo_c.describe()
```

```python
bj_meo_c.hist(bins=15, color='steelblue', edgecolor='black', linewidth=1.0,
              xlabelsize=8, ylabelsize=8, grid=False)  
plt.tight_layout(rect=(0, 0, 1.2, 1.2))
```

```python
c = ['station_id','longitude','latitude','utc_time']

result = pd.concat([bj_meo[c], bj_meo_c], axis=1)
result.head()
```

```python
result.to_csv(path+"process_data/bj_meo_c.csv", index=False)
```

```python
# Correlation Matrix Heatmap
f, ax = plt.subplots(figsize=(10, 6))
corr =result.corr()
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=.05)
f.subplots_adjust(top=0.93)
t= f.suptitle('Beijing observed meteorology Heatmap', fontsize=14)
```
